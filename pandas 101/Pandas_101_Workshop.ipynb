Let's discuss now, `.loc` and `iloc`

```python
#df[0] will not work

df.iloc[0] #PRINTS THE FIRST ROW
```

```python
df.iloc[441] #PRINTS THE 441st ROW
```

`iloc` specifies the index of the row (only accepts integers values)

whereas

`loc` specifies the label of the row (also accepts strings along with integer values)

```python
df.loc[0]
```

```python
df.loc[4, 'age']
```

```python
df.iloc[4, 0] # GET ME 4th VALUE IN FIRST COLUMN
```

```python
df.iloc[4, 0:2] #GET ME 4th RECORD OF FIRST TWO COLUMNS
```

```python
df.loc[4, ['age', 'sex']]
```

```python
df.iloc[4, [1,5,7,9]]
```

condition

```python
age_df_temp = df.loc[df.age < 5.38]
age_df_temp
```
